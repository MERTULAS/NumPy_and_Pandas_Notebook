# Reading Datasets From Files

```python
import pandas as pd
import numpy as np
```

## .csv Files

```python
### This .csv file downloaded from kaggle
dataset = pd.read_csv("Corona_NLP_test.csv")
dataset
```

```python
### Drop the first column (UserName)
newdataset = dataset.drop(["UserName"], axis=1)
newdataset
```

```python
### Saving the new dataset
newdataset.to_csv("newData_for_Corona.csv")
```

```python
### Saving the new dataset (without indexes)
newdataset.to_csv("newData_for_Corona_without_indexes.csv", index=False)
```

```python
dataset = pd.read_csv("newData_for_Corona_without_indexes.csv")
dataset
```
